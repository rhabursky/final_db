# Function to Extract Salary
Define a function that navigates the Payscale site: searches for the degree category, accesses the specific degree page, finds the salary link, and extracts the average base salary from the specified HTML element. Return 'N/A' if not found.

```python
base_url = "https://www.payscale.com"

program_to_category = {
    "Accounting": "Business-Management-and-Marketing",
    "Business Administration": "Business-Management-and-Marketing",
    "Finance": "Business-Management-and-Marketing",
    "Human Resource Management": "Business-Management-and-Marketing",
    "Information Systems": "Business-Management-and-Marketing",
    "International Business": "Business-Management-and-Marketing",
    "Management": "Business-Management-and-Marketing",
    "Marketing": "Business-Management-and-Marketing",
    "Sport Business Management": "Business-Management-and-Marketing",
    "Anthropology and Archaeology": "Social-Sciences",
    "Applied Sociology": "Social-Sciences",
    "Art": "Liberal-Arts",
    "Communication": "Communications",
    "Dance": "Liberal-Arts",
    "Economics": "Social-Sciences",
    "English": "Liberal-Arts",
    "History": "Liberal-Arts",
    "Journalism": "Communications",
    "Liberal Studies": "Liberal-Arts",
    "Music": "Liberal-Arts",
    "Music Education": "Liberal-Arts",
    "Philosophy": "Liberal-Arts",
    "Political Science": "Social-Sciences",
    "Psychology": "Social-Sciences",
    "Religious Studies": "Liberal-Arts",
    "Russian Studies": "Liberal-Arts",
    "Social Work": "Social-Sciences",
    "Sociology": "Social-Sciences",
    "Spanish and Spanish Education": "Liberal-Arts",
    "Studio Art": "Liberal-Arts",
    "Biochemistry": "Science",
    "Biology": "Science",
    "Chemistry & Biochemistry": "Science",
    "Computer Science": "Engineering",
    "Data Science": "Science",
    "Environmental Science": "Science",
    "Mathematics": "Science",
    "Neuroscience": "Science",
    "Physics": "Science",
    "Art Therapy": "Health-Care",
    "Exercise Science": "Health-Care",
    "Forensic Science": "Health-Care",
    "Nursing": "Health-Care",
    "Pre_Athletic Training": "Health-Care",
    "Public Health": "Health-Care",
    "Sports Medicine": "Health-Care",
    "Criminal Justice": "Criminal-Justice",
    "Fashion Merchandising": "Liberal-Arts",
    "Graphic Design": "Liberal-Arts",
    "Hospitality Management": "Hospitality-and-Tourism",
    "Intelligence Studies": "Liberal-Arts",
    "Sport and Event Management": "Liberal-Arts"
}

def extract_salary(program):
    if program not in program_to_category:
        return 'N/A'
    category_slug = program_to_category[program]
    category_url = base_url + "/research/US/Degree/" + category_slug
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    try:
        response = requests.get(category_url, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')
        # Find links to degrees
        links = soup.find_all('a', href=re.compile(r'/research/US/Degree=.*/Salary'))
        for link in links:
            text = link.get_text(strip=True)
            if program.lower() in text.lower():
                degree_url = base_url + link.get('href')
                # Fetch degree page
                response2 = requests.get(degree_url, headers=headers)
                response2.raise_for_status()
                soup2 = BeautifulSoup(response2.content, 'html.parser')
                salary_span = soup2.find('span', class_='default-overview__value')
                if salary_span:
                    return salary_span.get_text(strip=True)
                break
    except:
        pass
    return 'N/A'
```

# Scrape Salaries for Each Program
Loop through the program list, call the extraction function for each, and collect the results in a list of tuples (title, salary).

```python
results = []
for prog in programs:
    salary = extract_salary(prog)
    results.append((prog, salary))
    print(f"Processed {prog}: {salary}")
```

# Display Results
Print or display the list of tuples with program titles and their corresponding salaries.

```python
print(results)
```
